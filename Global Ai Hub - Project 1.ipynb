```python
import pandas as pd
students = {'Name': ['Hasan','Ümit','Fatih','Berfin','Umut','Deniz','Hayal','Sude','Emre','Avni'], 
            'Last_Name': ['Can','Ekmek','Yap','Gezici','Yakın','Uzak','Dogan','İyi','Sev','Menekse'],
            'School_No': [1339,1000,203,367,678,23,789,10,1001,210],
            'Math_Grade': [80, 49, 51, 63, 76, 39, 13, 100, 0, 69]}

df = pd.DataFrame(students)
print(df)

class math():

    def __init__(self,name,lastname,school_no,math_grade):
        self.name=name
        self.lastname=lastname
        self.school_no=school_no
        self.math_grade=math_grade
        
    def grade(self):
        df['Letter_Grade'] = ''
        for i in range(0,len(df)):
            if self.math_grade[i]>=80 and self.math_grade[i]<=100:
                df.loc[[i],['Letter_Grade']] = 'A'
            elif self.math_grade[i]>=70 and self.math_grade[i]<=79:
                df.loc[[i],['Letter_Grade']] = 'B'
            elif self.math_grade[i]>=60 and self.math_grade[i]<=69:
                df.loc[[i],['Letter_Grade']] = 'C'
            elif self.math_grade[i]>=50 and self.math_grade[i]<=59:
                df.loc[[i],['Letter_Grade']] = 'D'
            else:
                df.loc[[i],['Letter_Grade']] = 'F'
        print(df)

    def success(self):
            df['Result'] = ''
            for i in range(0,len(df)):
                if self.math_grade[i] < 50:
                    df.loc[[i],['Result']]= 'Couldnt Passed'
                else:
                    df.loc[[i],['Result']]= 'Passed'
            print(df)
            
            
            
obje = math(df['Name'], df['Last_Name'], df['School_No'], df['Math_Grade'])
obje.grade()
obje.success()
df.to_excel('Students_grades_result.xlsx')




    

```
